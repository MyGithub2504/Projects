# SCALING 

Scaling is not required here because all the data in columns not having huge difference between them.

```python
x=Mushroom_data.drop("class",axis=1)   #indeppendent varaibles 
y=Mushroom_data["class"]           #dependent varaiable
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

for i in range(200):
    x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.33, random_state=i)
    from sklearn.tree import DecisionTreeClassifier
    lm=DecisionTreeClassifier()
    lm.fit(x_train,y_train)
    pred=lm.predict(x_test)
    accu=accuracy_score(y_test,pred)
    print(accu*100, "at random_state", i)
```

```python
from sklearn.tree import DecisionTreeClassifier
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.30, random_state=i)
    mod=DecisionTreeClassifier()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ", maxAccu, "on Random_state", maxRS)
```

Best accuracy is 1 at random state 1

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.30, random_state=1)
```

Now we are checking accuracy using various algorithm.

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier


```

```python
rf=RandomForestClassifier(n_estimators=100,random_state=1)
#RandomForestClassifier(100)----> Default
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
#AdaBoostClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=50,learning_rate=1.0)
ad=AdaBoostClassifier()
ad.fit(x_train,y_train)
ad_pred=ad.predict(x_test)
print(accuracy_score(y_test,ad_pred))
print(confusion_matrix(y_test,ad_pred))
print(classification_report(y_test,ad_pred))
```

```python
# Import Support vector classifier
from sklearn.svm import SVC
SV = SVC()
SV.fit(x_train, y_train)
pred = SV.predict(x_test)
acc=classification_report(y_test, pred)
print(acc)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
gb_pred=gb.predict(x_test)
print(accuracy_score(y_test,gb_pred))
print(confusion_matrix(y_test,gb_pred))
print(classification_report(y_test,gb_pred))
```

Now we will check the cross validation score

```python
from sklearn.model_selection import cross_val_score
```

```python
print(cross_val_score(rf,x,y,cv=5).mean())
```

```python
print(cross_val_score(ad,x,y,cv=5).mean())
```

```python
print(cross_val_score(SV,x,y,cv=5).mean())
```

```python
print(cross_val_score(gb,x,y,cv=5).mean())
```

THE value with least difference will be the best model. here gb(Gradient Booster Classifier) has been

chosen as the best model for hyper parameter tuning

# Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {'loss':['deviance','exponential'],
            'n_estimators':np.arange(2,10), 
            'criterion':['friedman_mse','squared_error','mse','mae'],
            'max_depth' :np.arange(1,8)}
```

```python
GCV=GridSearchCV(GradientBoostingClassifier(),parameter,cv=5)
```

```python
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

```python


Final_model = GradientBoostingClassifier(criterion = 'friedman_mse', loss = 'deviance', max_depth = 7, n_estimators = 2)
Final_model.fit(x_train, y_train)
pred = Final_model.predict(x_test)
acc=accuracy_score(y_test, pred)
print(acc*100)
```

```python
import numpy as np 
from sklearn .metrics import roc_auc_score 
auc = np.round(roc_auc_score(y_test,pred), 3) 
print("AUC for our sample data is",(auc))
```

```python
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
AUC_ROC = metrics.roc_curve(y_test,pred)
plt.plot(AUC_ROC)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

Value of AUC-ROC shows that model is successful at separating different classes
