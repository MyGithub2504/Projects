# HyperParameter Tuning 

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter= {'n_estimators': np.arange(2,9),
              'learning_rate':np.arange(1,7),
            'subsample': np.arange(1,9),
           'criterion' :['friedman_mse','squared_error','mse','mae']}
 
```

```python
GCV = GridSearchCV(GradientBoostingClassifier(), parameter, cv=5) 
```

```python
GCV.fit(x_train, y_train)
```

```python
GCV.best_params_
```

```python
Final_model = GradientBoostingClassifier(criterion = 'squared_error', learning_rate = 1, n_estimators = 8, subsample=1)
Final_model.fit(x_train, y_train)
pred = Final_model.predict(x_test)
f1score=f1_score(y_test, pred)
print(f1score*100)
```

### AUC-ROC Curve

```python
from sklearn.metrics import roc_curve, auc
fpr,tpr,thresholds=roc_curve(pred,y_test)
roc_auc=auc(fpr,tpr)
```

```python
plt.figure()
plt.plot(fpr,tpr,color='darkorange',lw=10,label='ROC curve(area=%0.2f)'% roc_auc)
plt.plot([0,1],[0,1],color='navy', lw=10, linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
