```python
from sklearn.preprocessing import power_transform
x=power_transform(Titanicx, method='yeo-johnson')
x
```

```python
from sklearn.preprocessing import StandardScaler
SC=StandardScaler()
X=SC.fit_transform(x)
X
```

```python
Y=Titanicy
```

# train_test_split

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

for i in range(200):
    x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.33, random_state=i)
    from sklearn.tree import DecisionTreeClassifier
    lm=DecisionTreeClassifier()
    lm.fit(x_train,y_train)
    pred=lm.predict(x_test)
    f1score=f1_score(y_test,pred)
    print("fiscore is",f1score, "at random_state", i)
```

```python
from sklearn.tree import DecisionTreeClassifier
maxf1=0
maxRS=0
for i in range(1,200):
    x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.33, random_state=i)
    mod=DecisionTreeClassifier()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    f1score=f1_score(y_test,pred)
    if f1score>maxf1:
        maxf1=f1score
        maxRS=i
print("Best accuracy is ", maxf1, "on Random_state", maxRS)
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.33, random_state=5)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
```

```python
rf=RandomForestClassifier(n_estimators=100,random_state=5)
#RandomForestClassifier(100)----> Default
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(f1_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
#AdaBoostClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=50,learning_rate=1.0)
ad=AdaBoostClassifier()
ad.fit(x_train,y_train)
ad_pred=ad.predict(x_test)
print(f1_score(y_test,ad_pred))
print(confusion_matrix(y_test,ad_pred))
print(classification_report(y_test,ad_pred))
```

```python
# Import Support vector classifier
from sklearn.svm import SVC
SV = SVC()
SV.fit(x_train, y_train)
pred = SV.predict(x_test)
classreport=classification_report(y_test, pred)
print(classreport)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
gb_pred=gb.predict(x_test)
print(f1_score(y_test,gb_pred))
print(confusion_matrix(y_test,gb_pred))
print(classification_report(y_test,gb_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)
neigh_pred=neigh.predict(x_test)
print(f1_score(y_test,neigh_pred))
print(confusion_matrix(y_test,neigh_pred))
print(classification_report(y_test,neigh_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter={'degree': np.arange(1,8),
           'coef0' : np.arange(1,6), 
           'tol' : np.arange(2,8),
           'cache_size':np.arange(2,5)
}
```

```python
GCV = GridSearchCV(SVC(), parameter, cv=5) 
```

```python
GCV.fit(x_train, y_train)
```

```python
GCV.best_params_
```

```python
Final_model = SVC(cache_size = 2, coef0 = 1,  degree=1, tol=2)
Final_model.fit(x_train, y_train)
pred = Final_model.predict(x_test)
f1score=f1_score(y_test, pred)
print(f1score*100)
```

## AUC-ROC Curve

```python
from sklearn.metrics import roc_curve, auc
fpr,tpr,thresholds=roc_curve(pred,y_test)
roc_auc=auc(fpr,tpr)
```

```python
plt.figure()
plt.plot(fpr,tpr,color='darkorange',lw=10,label='ROC curve(area=%0.2f)'% roc_auc)
plt.plot([0,1],[0,1],color='navy', lw=10, linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
